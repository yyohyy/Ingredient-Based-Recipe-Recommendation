```python
import re 

def remove_noise(ingredients):
    quantity = r"(\d+|\d+\s?/\s?\d+)"
    measurements = (
        r"(tsp|tbsp|cup|cups|ml|l|kg|g|grams|oz|ounce|pound|lb|"
        r"pinch|dash|handful|clove|slice|can|packet|sprig|stalk|pack"
        r"bunch|teaspoon|tablespoon|litre|litres|quart|pint|gal|sachet)"
    )
    adjectives = r"\b(strong|fine|white|red|fresh|large|small|good-quality)\b"
    cleaned_ingredients = []
    for ingredient in ingredients:
        ingredient = ingredient.strip().lower()
        ingredient = re.sub(quantity, "", ingredient)
        ingredient = re.sub(rf"\b{measurements}\b", "", ingredient)
        ingredient = re.sub(rf"\b{adjectives}\b", "", ingredient)
        ingredient = re.sub(r"\s+", " ", ingredient)
        ingredient = ingredient.split(',', 1)[0].strip()
        cleaned_ingredients.append(ingredient)
    return cleaned_ingredients
```

```python
cleaned_ingredients = remove_noise(combined_dataset['ingredients'].iloc[0])
print(cleaned_ingredients)
```

```python
combined_dataset['cleaned_ingredients'] = combined_dataset['ingredients'].apply(remove_noise)
```

# Content Based Filtering 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer(stop_words='english')
ingredient_matrix = tfidf.fit_transform(combined_dataset['cleaned_ingredients'].apply(lambda x: ' '.join(x)))
content_similarity = cosine_similarity(ingredient_matrix)
```

```python
def find_recipes(input_ingredients, top_n=5):
    input_ingredients_str = ' '.join(input_ingredients)

    input_vector = tfidf.transform([input_ingredients_str])
    
    similarity_scores = cosine_similarity(input_vector, ingredient_matrix).flatten()
    
    similar_indices = similarity_scores.argsort()[-top_n:][::-1]
    
    similar_recipes = combined_dataset.iloc[similar_indices][['name', 'description','ingredients','steps']]
    similar_recipes['similarity_score'] = similarity_scores[similar_indices]
    return similar_recipes
```

```python
input_ingredients = ['chicken','rice','potato']
similar_recipes = find_recipes(input_ingredients)
print(similar_recipes)
```
